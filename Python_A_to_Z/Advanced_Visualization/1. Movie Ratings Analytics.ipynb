```python
list1 = list()  #criar uma lista vazia
mylabels = list()
for gen in movies.Genre.cat.categories:     
    list1.append(movies[movies.Genre == gen].BudgetMillions)  #appending the series, ou seja, adicionar algo do lado direito da lista
    mylabels.append(gen)

h = plt.hist(list1, bins=30, stacked=True, rwidth=1, label=mylabels)
plt.legend()
plt.show()
```

```python
#Thematic Edits
```

```python
list1 = list()
mylabels = list()
for gen in movies.Genre.cat.categories:     
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)

sns.set_style("whitegrid")
fig, ax = plt.subplots()   #fazemos isto para nos permitir mexer no tamanho do gr√°fico livremente
fig.set_size_inches(11.7,8.27)  #size of an A4
h = plt.hist(list1, bins=30, stacked=True, rwidth=1, label=mylabels)
plt.title("Movie Budget Distribution", fontsize=35, color='DarkBlue', fontname="Console")
plt.ylabel("Number of Movies", fontsize=25, color="Red")
plt.xlabel("Budget", fontsize=25, color="Green")
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(frameon=True,fancybox=True,shadow=True, framealpha=1, prop={'size':20})
plt.show()
```
